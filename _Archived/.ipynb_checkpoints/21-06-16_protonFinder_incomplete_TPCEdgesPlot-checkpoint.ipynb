```python
%%time
reload(cl)
reload(ll)

### FILENAMES
cpath = '/home/dporzio/Data/Michel/michel_datalog_2021_04_04_16_19_13_CEST_evd.h5'
ldir = '/home/dporzio/Data/Michel'

### LOAD LIGHT DATA
# Uproot
ldata, lpath = ml.FindPartnerLightFile(cpath,ldir)
# Root
rfile = ROOT.TFile.Open(lpath, 'read')
rwf = rfile.Get('rwf')
# Convert utime to a better usable format
ldata['utime_ms'] = ldata['utime_ms'].astype(np.int64)

### LOAD CHARGE DATA
cdata = h5py.File(cpath,'r')

h_events = cdata['events']
h_trigs = cdata['ext_trigs']
h_tracks = cdata['tracks']
h_hits = cdata['hits']
h_info = cdata['info'].attrs

print("events:",[t for t in cdata['events'].dtype.names])
print("tracks:",[t for t in cdata['tracks'].dtype.names])
print("hits:",[t for t in cdata['hits'].dtype.names])
print("ext_trigs:",[t for t in cdata['ext_trigs'].dtype.names])
print("info:",[t for t in (cdata['info'].attrs).keys()])

### LOAD GEOMETRY DATA
geometryHelper = cl.GetGeometryHelper(h_info)

### OTHER FUNCTIONS
def EuDistance(p1,p2):
    return np.sqrt(np.power(p1[0]-p2[0],2.) + np.power(p1[1]-p2[1],2.) + np.power(p1[2]-p2[2],2.))
```

```python
### CREATE LIST OF GOOD EVENTS
# Consider good events all events that have max 15 tracks
# And that have an external trigger
sel_maxTracksInEv = 20

bool_tracks = h_events['ntracks'] < sel_maxTracksInEv
bool_trigger = h_events['n_ext_trigs'] > 0
goodEvs = h_events[(bool_tracks & bool_trigger)]
```

```python
nSample = 100000
nbins = 80

dim = 0
start = [track['start'][dim] for track in h_tracks[0:nSample]]
end = [track['end'][dim] for track in h_tracks[0:nSample]]
x_ext = start+end

dim = 1
start = [track['start'][dim] for track in h_tracks[0:nSample]]
end = [track['end'][dim] for track in h_tracks[0:nSample]]
y_ext = start+end

dim = 2
start = [track['start'][dim] for track in h_tracks[0:nSample]]
end = [track['end'][dim] for track in h_tracks[0:nSample]]
z_ext = start+end
```

```python

reload(bbl)
reload(pcc)
sel_minTrackLength = 2 # cm. Minimum length of proton candidate track
sel_maxTrackLength = 50 # cm. Maximum length of proton candidate track
sel_dedxAveBin = 1 # cm. Length of bin to add together dEdx
sel_mindedx_end = 2 # MeV/cm. Minimum dedx at the end of proton candidate track
sel_maxdedx_start = 1 # MeV/cm. Maximum dedx at the start of proton candidate track
conversionFactor = 245 # #e-/mv, q is in mv, which corresponds to 245 electrons
rmodel = 'box' # Recombination model


protons = []
# Loop through each event of the good event list
for ev in tqdm(goodEvs[0:10000]):
    # For each event, get a list of all the tracks
    tracks = h_tracks[ev['track_ref']]
    # Loop through each track
    for track in tracks:        
        # Let's get event id and track id and create the candidate
        evid = ev['evid']
        trid = track['track_id']
        proton = pcc.ProtonCandidate(evid,trid)
        
        # First of all get the tref which we need for z-position
        proton.ev_tref = cl.GetEventStartTime(evid,cdata)

        # Apply some early selections on length alone
        length = track['length']/10
        if length<sel_minTrackLength or length>sel_maxTrackLength:
            continue
        # And also on containment
        
        
        # We get the edge points, we don't make assumptions on which is start and which is end
        p0 = track['start']
        p1 = track['end']
        hits = h_hits[track['hit_ref']]
        
        # Just for now, assume end is real end
        # Let's order all hits in term of distance from end
        distances_fromEnd, distances_fromStart, n_els, charges = [], [], [], []
        for hit in hits:
            # z-position we have to calculate ourselves. Annoying, but that's the way it is
            hz = cl.ConvertTimeToZ(geometryHelper,hit['iogroup'],hit['iochannel'],hit['ts'],proton.ev_tref)
            hitLoc = [hit['px'],hit['py'],hz]
            # Calculate distance from end, IN CM
            distances_fromEnd.append(EuDistance(p1,hitLoc)/10.)
            distances_fromStart.append(EuDistance(p0,hitLoc)/10.)
            n_el = hit['q']*conversionFactor
            charge = bbl.ConvertQtoE(n_el,model=rmodel)
            n_els.append(n_el)
            charges.append(charge)
        # Numpify all the arrays we made
        distances_fromEnd = np.array(distances_fromEnd)
        distances_fromStart = np.array(distances_fromStart)
        n_els = np.array(n_els)
        charges = np.array(charges)
        
        # Find max distance, that's also length of track
        maxDistance = distances_fromEnd.max()
        minDistance = distances_fromEnd.min()
        # Calculate dedx at start and end in order to judge if track is flipped
        dedxSel_end = (distances_fromEnd>0) & (distances_fromEnd<sel_dedxAveBin)
        dedxSel_start = (distances_fromEnd>maxDistance-sel_dedxAveBin) & (distances_fromEnd<maxDistance)
        dedx_end = (charges[dedxSel_end]).sum()/sel_dedxAveBin
        dedx_start = (charges[dedxSel_start]).sum()/sel_dedxAveBin
        
        # Another condition, let's continue with the candidate only if one of the extremities
        # has sufficient dedx for proton
        cond_gooddedx_end = max(dedx_end,dedx_start) > sel_mindedx_end
        cond_gooddedx_start = min(dedx_end,dedx_start) < sel_maxdedx_start
        cond_gooddedx = cond_gooddedx_end and cond_gooddedx_start
#         print(dedx_end,dedx_start)
        
        if not cond_gooddedx:
            continue
        
        # If we get past this, good news, this might be a proton
        # Let's check if we have to flip it
        
        # If the dedx is inversed we assume the track is flipped
        # and we recalculate dedx and range from actual end point
        if dedx_end<dedx_start:
            proton.dir_flipped = True
            distances = distances_fromStart   
            proton.pos_xStart = track['end'][0]
            proton.pos_yStart = track['end'][1]
            proton.pos_zStart = track['end'][2]
            proton.pos_tStart = track['end'][3]
            proton.pos_xEnd = track['start'][0]
            proton.pos_yEnd = track['start'][1]
            proton.pos_zEnd = track['start'][2]
            proton.pos_tEnd = track['start'][3]
        else:
            proton.dir_flipped = False
            distances = distances_fromEnd
            proton.pos_xStart = track['start'][0]
            proton.pos_yStart = track['start'][1]
            proton.pos_zStart = track['start'][2]
            proton.pos_tStart = track['start'][3]
            proton.pos_xEnd = track['end'][0]
            proton.pos_yEnd = track['end'][1]
            proton.pos_zEnd = track['end'][2]
            proton.pos_tEnd = track['end'][3]
        
        # Store the raw dedx values
        ind_sort = np.argsort(distances)
        proton.dedx_distances = distances[ind_sort]
        proton.dedx_charges = charges[ind_sort]
        
        # Calculate the binned dedx
        proton.dedx_rbins = []
        proton.dedx_rbins_mean = []
        proton.dedx_qbins = []
        pidas = []
        start_rbins = np.arange(0,101,sel_dedxAveBin)
        for i in range(len(start_rbins)-1):
            thisBin = start_rbins[i]
            nextBin = start_rbins[i+1]
            thisMeanBin = (thisBin + nextBin)/2.
            selected_bins = (proton.dedx_distances>=thisBin) & (proton.dedx_distances<nextBin)
            sum_charges = (proton.dedx_charges[selected_bins]).sum()  
            thisPida = sum_charges * np.power(thisMeanBin,0.42)
            
            proton.dedx_rbins.append(thisBin)
            proton.dedx_rbins_mean.append(thisMeanBin)
            proton.dedx_qbins.append(sum_charges)
            if thisMeanBin<30 and thisPida!=0:
                pidas.append(thisPida)
        proton.pidaScore = (np.array(pidas)).mean()
            
        protons.append(proton)
        
allPidas = [protons[i].pidaScore for i in range(len(protons))]
```

```python
nbins = 100
xlims = [0,20]

plt.figure(figsize=(8,5),facecolor='white',dpi=100)
hp.HepHist(allPidas,nbins=nbins,xlims=xlims,color='C0',label='Corr.')
plt.grid(ls='--',color='C7',alpha=0.1)
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", borderaxespad=0, frameon=False)
plt.show()
```
