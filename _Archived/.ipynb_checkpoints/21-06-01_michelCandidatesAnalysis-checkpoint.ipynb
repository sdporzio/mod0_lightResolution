```python
%%time
reload(cl)
reload(ll)

### FILENAMES
cpath = '/home/dporzio/Data/Michel/michel_datalog_2021_04_04_16_19_13_CEST_evd.h5'
ldir = '/home/dporzio/Data/Michel'

### LOAD LIGHT DATA
# Uproot
ldata, lpath = ml.FindPartnerLightFile(cpath,ldir)
# Root
rfile = ROOT.TFile.Open(lpath, 'read')
rwf = rfile.Get('rwf')
# Convert utime to a better usable format
ldata['utime_ms'] = ldata['utime_ms'].astype(np.int64)

### LOAD CHARGE DATA
cdata = h5py.File(cpath,'r')

h_events = cdata['events']
h_trigs = cdata['ext_trigs']
h_tracks = cdata['tracks']
h_hits = cdata['hits']
h_info = cdata['info'].attrs

print("events:",[t for t in cdata['events'].dtype.names])
print("tracks:",[t for t in cdata['tracks'].dtype.names])
print("hits:",[t for t in cdata['hits'].dtype.names])
print("ext_trigs:",[t for t in cdata['ext_trigs'].dtype.names])
print("info:",[t for t in (cdata['info'].attrs).keys()])

### LOAD GEOMETRY DATA
geometryHelper = cl.GetGeometryHelper(h_info)

# LOAD GOOD MICHELS
filename = '/home/dporzio/App/MichelElectrons_Analysis/GoodMichels/goodMichels.pkl'
df = pickle.load(open(filename,'rb'))

def GetOrientation(start,end):
    dx = end[0] - start[0]
    dy = end[1] - start[1]
    dz = end[2] - start[2]
    theta = np.arctan2(np.linalg.norm([dx,dy]),dz)
    phi = np.arctan2(dy,dx)
    return [dx,dy,dz], theta, phi

def EuDistance(p1,p2):
    return np.sqrt(np.power(p1[0]-p2[0],2.) + np.power(p1[1]-p2[1],2.) + np.power(p1[2]-p2[2],2.))

# reload(evd)
# reload(cl)
# evd.SmallEventDisplay(15,cpath,cdata,geometryHelper,rot=45,showTracks=True,xTrackOffset=100)
# plt.show()
```

```python
i = 20
reload(evd)
reload(cl)
# evd.SmallEventDisplay(df['id_eventId'].iloc[i],cpath,cdata,geometryHelper,rot=45,showTracks=True,xTrackOffset=100)
# plt.show()

rrbins = np.linspace(0,80,20)
stepPerCm = (rrbins[1]-rrbins[0])
conversionFactor = 245 # #e-/mv, q is in mv, which corresponds to 245 electrons
rrvals_tot = np.zeros(len(rrbins)-1)
rrweights = np.zeros(len(rrbins)-1)

for g in tqdm(range(len(df))):
    michel = df.iloc[g]

    rawcharge, charge, distance = [], [], []
    tref = cl.GetEventStartTime(michel['id_eventId'],cdata)

    for tId in michel['stitch_iStitches']:
        hits = h_hits[h_tracks[tId]['hit_ref']]
        for hit in hits:
            rawcharge.append(hit['q_raw'])
            charge.append(hit['q'])
            hz = cl.ConvertTimeToZ(geometryHelper,hit['iogroup'],hit['iochannel'],hit['ts'],tref)
            hitLoc = [hit['px'],hit['py'],hz]
            distance.append(EuDistance(michel['mu_pos_end'],hitLoc)/10.)
    resRange = pd.DataFrame()
    resRange['charge'] = charge
    resRange['rawcharge'] = rawcharge
    resRange['distance'] = distance
    
    rrvals = np.array([resRange.query(f'distance>={rrbins[i]} & distance<{rrbins[i+1]}')['charge'].sum() for i in range(len(rrbins)-1)])
    maxDist = resRange['distance'].max()
    rrweights = rrweights+(rrbins[:-1]<maxDist)
    rrvals_tot = rrvals_tot + rrvals
```

```python

reload(bbl)
mean_dist_box, mean_dqdx_box, mpv_dist_box, mpv_dqdx_box = bbl.ResRange(105.65,outType='dqdx',recModel='box')
mean_dist_birks, mean_dqdx_birks, mpv_dist_birks, mpv_dqdx_birks = bbl.ResRange(105.65,outType='dqdx',recModel='birks')
    
# Convert rrvals
rrvals_tot_c = (rrvals_tot*conversionFactor/stepPerCm)/rrweights
plt.figure(figsize=(8,5),facecolor='white',dpi=100)
hp.HepPlot(rrbins[:-1],rrvals_tot_c,color=mc[0],label='Stopping muons (data)')
plt.plot(mean_dist_birks,mean_dqdx_birks,color='C1',lw=3,label='$\mu$ Mean (Birks)')
plt.plot(mpv_dist_birks,mpv_dqdx_birks,color='C1',ls='--',lw=3,label='$\mu$ MPV (Birks)')
plt.plot(mean_dist_box,mean_dqdx_box,color='C3',lw=3,label='$\mu$ Mean (Mod.Box)')
plt.plot(mpv_dist_box,mpv_dqdx_box,color='C3',ls='--',lw=3,label='$\mu$ MPV (Mod.Box)')
plt.grid(True,ls='--',color='C7',alpha=0.1)
plt.xlabel('Residual range [cm]')
plt.ylabel('dQ/dx [e$^{-}/cm$]')
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", borderaxespad=0, frameon=False)
plt.ylim(0,250e3)
plt.xlim(0,80)
plt.show()
```

```python
reload(evd)
reload(cl)
reload(bbl)

rrbins = np.linspace(0,80,20)
cmPerBin = (rrbins[1]-rrbins[0])
rrmeans = [(rrbins[i]+rrbins[i+1])/2. for i in range(len(rrbins)-1)]
rrvals_tot = np.zeros(len(rrbins)-1)
rrweights = np.zeros(len(rrbins)-1)
mean_pida = []

for g in tqdm(range(len(df))):
    michel = df.iloc[g]

    rawcharge, charge, distance = [], [], []
    tref = cl.GetEventStartTime(michel['id_eventId'],cdata)

    for tId in michel['stitch_iStitches']:
        hits = h_hits[h_tracks[tId]['hit_ref']]
        for hit in hits:
            rawcharge.append(hit['q_raw'])
            charge.append(hit['q'])
            hz = cl.ConvertTimeToZ(geometryHelper,hit['iogroup'],hit['iochannel'],hit['ts'],tref)
            hitLoc = [hit['px'],hit['py'],hz]
            distance.append(EuDistance(michel['mu_pos_end'],hitLoc)/10.)
    resRange = pd.DataFrame()
    resRange['charge'] = charge
    resRange['rawcharge'] = rawcharge
    resRange['distance'] = distance
    
    rrvals = np.array([resRange.query(f'distance>={rrbins[i]} & distance<{rrbins[i+1]}')['charge'].sum() for i in range(len(rrbins)-1)])*conversionFactor/cmPerBin
    evals = [bbl.ConvertQtoE(rrvals[i],model='box') for i in range(len(rrvals))]
    pida = []
    for i in range(len(rrbins)-1):
        if rrmeans[i]<30:
            pida.append(evals[i]*np.power(rrmeans[i],0.42))
    pida = np.array(pida)
    mean_pida.append(np.mean(pida))
    
    maxDist = resRange['distance'].max()
    rrweights = rrweights+(rrbins[:-1]<maxDist)
    rrvals_tot = rrvals_tot + rrvals
```

```python
clean_pida = [x for x in mean_pida if str(x) != 'nan']
```

```python
# Convert rrvals
plt.figure(figsize=(8,5),facecolor='white',dpi=100)
hp.HepHist(clean_pida,color=mc[0],nbins=20,xlims=[0,20],label='Stopping muons (data)')
plt.grid(True,ls='--',color='C7',alpha=0.1)
plt.xlabel('PIDA [MeV/cm$^{1.42}$]')
plt.ylabel('A.U.')
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", borderaxespad=0, frameon=False)
# plt.ylim(0,250e3)
plt.xlim(0,20)
plt.show()
```
